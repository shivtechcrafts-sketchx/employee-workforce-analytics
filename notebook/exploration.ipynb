# Employee Data Exploration (EDA)

Objective: Understand the dataset, identify issues, and extract insights before cleaning and modeling.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../data/raw_data.csv")
df.head()
```

```python
print("Shape:", df.shape)
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].unique())
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Salary'])
plt.title("Salary Outliers")
plt.show()
```

```python
sns.histplot(df['Age'], bins=10)
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x='City', y='Salary', data=df)
plt.title("Average Salary by City")
plt.show()
```

```python
sns.barplot(x='Gender', y='Salary', data=df)
plt.title("Salary by Gender")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

## Key Insights from EDA

1. Most employees are between 2035 years old.
2. Salary varies significantly across cities.
3. Some missing values exist in Age and Salary.
4. Few outliers are present in salary.
5. Gender-based salary difference is visible.

These insights guided the data cleaning and modeling steps.
